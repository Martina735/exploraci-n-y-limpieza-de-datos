```python
import numpy as np
import pandas as pd
```

```python
rng=np.random.default_rng()
a_values=rng.integers(low=0, high=100, size=10)
b_values=rng.integers(low=100, high=500, size=10)
data = pd.DataFrame({"a":a_values, "b": b_values})
data
```

```python
data["a_cat"]=pd.cut(data.a, [0,30,40,50,60,100], labels=["muy poco", "poco", "normal", "mucho", "demasiado"])
data
```

```python
a_cat_mapper={"poco":"insuficiente",
"muy poco":"insuficiente", 
"normal":"suficiente",
"mucho":"suficiente",
"demasiado":"suficiente"}
```

```python
data["a_cat_suf_insuf"]=data.a_cat.map(a_cat_mapper)
data
```

```python
a_categories_1_dummies = pd.get_dummies(data.a_cat, drop_first=True,prefix="a_cat1")
a_categories_1_dummies
```

```python
a_categories_1_dummies = pd.get_dummies(data.a_cat, drop_first=True,prefix="a_cat1",dtype=int)
a_categories_1_dummies
```

```python
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")
print(penguins.shape)
penguins.sample(2)
```

```python
penguins.pivot_table("body_mass_g", index="species", columns="sex")
```

```python
penguins["flipper_length_mm"].describe()[["min","mean", "max"]]
```

```python
flipper_categories = pd.cut(penguins.flipper_length_mm, [172,201,231])
penguins.pivot_table("body_mass_g", ["species", flipper_categories], "sex")
```

```python
bill_categories = pd.qcut(penguins.bill_length_mm,3,labels=["chico", "medio", "grande"])
penguins.pivot_table("body_mass_g",["species", flipper_categories],[bill_categories,"sex"], fill_value="no existe" )
```
